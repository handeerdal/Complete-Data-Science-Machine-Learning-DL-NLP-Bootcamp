```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Load the data
df=sns.load_dataset('tips')
```

```python
df
```

```python
X=df.drop('total_bill',axis=1)
y=df['total_bill']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#import label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le1=LabelEncoder()
le2=LabelEncoder()
X_train['sex']=le.fit_transform(X_train['sex'])
X_train['smoker']=le1.fit_transform(X_train['smoker'])
X_train['time']=le2.fit_transform(X_train['time'])

X_test['sex']=le.transform(X_test['sex'])
X_test['smoker']=le1.transform(X_test['smoker'])
X_test['time']=le2.transform(X_test['time'])

```

```python
#import ohe
X_train = pd.get_dummies(X_train, columns=["day"], drop_first=False)
X_train[['day_Fri', 'day_Sat', 'day_Sun', 'day_Thur']] = X_train[['day_Fri', 'day_Sat', 'day_Sun', 'day_Thur']].astype(int)
X_test = pd.get_dummies(X_test, columns=["day"], drop_first=False)
X_test[['day_Fri', 'day_Sat', 'day_Sun', 'day_Thur']] = X_test[['day_Fri', 'day_Sat', 'day_Sun', 'day_Thur']].astype(int)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#import SVR
from sklearn.svm import SVR
svr=SVR()
svr.fit(X_train,y_train)
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score 
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
